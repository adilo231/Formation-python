# Structure de données (Listes et Tuples)

```python


def fibonacci(n):
    a = 0
    b = 1
    fib = [a] # Création d'une liste fib 
    while b < n:
        a, b = b, a+b
        fib.append(a) # ajoute la nouvelle valeur de a a la fin de fib
    return fib

print(fibonacci(1000))
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn import svm

# Generate a simple 2D dataset
X, y = datasets.make_blobs(n_samples=100, centers=2, random_state=3)

# Create a SVM classifier
clf = svm.SVC(kernel='linear', C=0.01)
clf.fit(X, y)



# Get the support vectors and their indices
support_indices = clf.support_
support_vectors = X[support_indices]


# Get the coefficients of the hyperplane (w) and the intercept (b)
w = clf.coef_[0]
b = clf.intercept_[0]
print(clf.coef_)
# Calculate the slope and intercept of the hyperplane
slope = -w[0] / w[1]
intercept = -b / w[1]

# Calculate the margin
margin = 1 / np.linalg.norm(w)

# Create a range of x values for plotting
xx = np.linspace(X[:, 0].min() - 1, X[:, 0].max() + 1, 100)

# Calculate the corresponding y values for the hyperplane
yy = slope * xx + intercept

# Calculate the parallel lines for the margin
yy_margin_plus = yy + margin
yy_margin_minus = yy - margin
plt.figure(figsize=(7,7))
# Plot the data points
plt.scatter(X[:, 0], X[:, 1], c=y, cmap=plt.cm.Paired)

# Plot the hyperplane and margins
plt.plot(xx, yy, 'k-')
plt.plot(xx, yy_margin_plus, 'k--')
plt.plot(xx, yy_margin_minus, 'k--')

# Plot the support vectors
plt.scatter(support_vectors[:, 0], support_vectors[:, 1], s=100, facecolors='none', edgecolors='k')



# Calculate the coordinates for vector w
w_normalized = w / np.linalg.norm(w)
b_coordinates = -b * w_normalized




plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.savefig('svm3.pdf')
plt.show()
```
